```python
def lookup_team_id_abv(team_name):
    for k,v in team_ids_names.items():
        for value in v[1]:
            if value == team_name:
                return k
    else:
        print(team_name+' is not a valid team name.')
        return 0
```

```python
data_dir = 'Data/teamstandings_basketballref/'
# yr = 2000
# f_name = 'standing_'+str(yr)+'.csv'
# standings = pd.read_csv(data_dir+f_name, skiprows=[0])
# standings['From'] = str(yr-1)+'-10-01'
# standings['To'] = str(yr)+'-06-30'

team_standings = pd.DataFrame()
yr = 1999 # starting year (data available from 1985 - 2016)
for i in range(0,19):
    f_name = 'standing_'+str(yr)+'.csv'
    standings = pd.read_csv(data_dir+f_name, skiprows=[0])
    standings['From'] = str(yr-1)+'-10-01'
    standings['To'] = str(yr)+'-06-30'
    team_standings = team_standings.append(standings)
    yr += 1
    
team_standings['Team_ID'] = team_standings['Team'].apply(lookup_team_id_abv)
```

```python
nba_data['HomeLastWin'] = False
nba_data['VisitorLastWin'] = False
won_last = defaultdict(int)
nba_data['HomeTeamWonLast'] = 0
last_match_winner = defaultdict(int)
nba_data['HomeTeamRanksHigher'] = 0

for index, row in nba_data.iterrows():
    date, home_team, visitor_team = row['Date'], row['Home_Team_ID'], row['Visitor_Team_ID']
    row['HomeLastWin'] = won_last[home_team]
    row['VisitorLastWin'] = won_last[visitor_team]
    nba_data.loc[index, ['HomeLastWin', 'VisitorLastWin']] = [row['HomeLastWin'], row['VisitorLastWin']]
    won_last[home_team] = row['HomeWin']
    won_last[visitor_team] = not row['HomeWin']
    ##
    teams = tuple(sorted([home_team, visitor_team]))
    row['HomeTeamWonLast'] = 1 if last_match_winner[teams] == row['Home_Team_ID'] else 0
    #nba_data.set_value(index, 'HomeTeamWonLast', row['HomeTeamWonLast'])
    nba_data.loc[index, 'HomeTeamWonLast'] = row['HomeTeamWonLast']
    winner = row['Home_Team_ID'] if row['HomeWin'] else row['Visitor_Team_ID']
    last_match_winner[teams] = winner
    ##
    home_rank = team_standings[(team_standings['Team_ID'] == home_team) & (team_standings['From']<date) & (team_standings['To']>date)]['Rk'].values[0]
    visitor_rank = team_standings[(team_standings['Team_ID'] == visitor_team) & (team_standings['From']<date) & (team_standings['To']>date)]['Rk'].values[0]
    row['HomeTeamRanksHigher'] = int(home_rank > visitor_rank)
    #nba_2014.set_value(index, 'HomeTeamRanksHigher', row['HomeTeamRanksHigher'])
    nba_data.loc[index, 'HomeTeamRanksHigher'] = row['HomeTeamRanksHigher']
```
